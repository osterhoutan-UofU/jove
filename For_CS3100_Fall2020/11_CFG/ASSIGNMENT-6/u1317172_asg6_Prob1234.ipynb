## Q3: Explain the workings of all the mk_X_nfa functions by arguing that they are making the right output NFA from the input NFA. 

In your explanation, explain how $q$ (`Q`), $\Sigma$ (`Sigma`), $\delta$ (`Delta`), $q_0$ (`q0`), and $F$ (`F`) are formed.
 
> 
### `mk_plus_nfa(N1, N2)`
_Given two NFAs, 
  return their union._
This function just unions all of the set containers of the NFAs (`N1` & `N2`).
It technically does this while creatign a new NFA object, 
  but that's not all that consiquiential,
  unless you really don't want to keep copies of the old NFA's in memory.
It then returns this new unioned NFA.



$\phantom{.}$
>
### `mk_cat_nfa(N1, N2)`
_Given two NFAs, 
  return their concatination._
Creates a new NFA that is the concatination of the NFA `N1` followed by the NFA `N2`.
Follows the same algorithm we learned way back when, 
  of connecting each final state of `N1` to each initial state of `N2`,
  by createing additional $\varepsilon$ paths
  (being carefull to not remove any $\varepsilon$ paths that the final state already had).
Making all of the initial states from `N2` non-initial states,
  and making all of the final states from `N1` non-final states.
To be careful about preserving data and not adding unnessisary states,
  this function acctualy creates a new NFA to return
  (just like technically all of the functions I'm discribing here).
That just unions the sets of states from `N1` & `N2` into the new NFA's $q$ set,
  it does the same thing for the alphabet sets ($\Sigma$),
  and it technically does the same thing for the $\delta$ function, 
  but earlier in the function before connecting `N1` and `N2`.
It then returns this NFA.



$\phantom{.}$
>
### `mk_star_nfa(N)`
_Given an NFA, 
  make its klien star._
Modifies an existing NFA `N`,
  so that whatever would have been accepted by the NFA can be repeated
  and make sure that $\varepsilon$ will be accepted as well.
To do this it craetes a new state that is both initial and final (`IF`),
  and at the end it will assign it to be the only initial and final state `N` has.
It creates a path from `IF` to the set of `N`'s old starting states 
  for the $\varepsilon$ condition.
It then makes sure that all of `N`'s old final states have $\varepsilon$ paths
  leading back to the `IF` state
  (making sure that all the previouse final states are not dead ends, 
    since they will no longer be final states).
It makes sure that our new `IF` state is a registered state in the NFA,
  and returns this NFA.



$\phantom{.}$
>
### `mk_eps_nfa()`
_An nfa with exactly one start+final state, 
  which is the NFA for Epsilon._
Creates a simple single state NFA.
Where the initial state is the final state,
  and no paths are assigned to the $\delta$ function.
Then return this simple NFA.


$\phantom{.}$
>
### `mk_symbol_nfa(a)`
_The NFA for a single RE letter._
Creates a simple 2 state NFA.
Where there is one initail state and one final state 
  (that are not the same state).
It then creates a single path for the NFA's $\delta$ function,
  that is from the initial state to the final state if the input matches `a`
  (the string that should be a single letter by teh standars of the RE).
Then return this simple NFA.
